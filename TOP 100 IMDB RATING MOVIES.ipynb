```python
import bs4
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc"
r = requests.get(url)# r variable has all the HTML code
htmlContent = r.content# r returns response so if we want the code we write r.content
#print(htmlContent)
```

```python
htmlText = r.text
#print(htmlText)
```

```python
soup = BeautifulSoup(r.content, 'html.parser')

#print(soup.prettify())	# to print html in tree structure
```

```python
title = soup.title
print(title)
```

```python
Movies_Name=[]
IMDB_Rating=[]
Year_Of_Release=[]
```

```python
Name=soup.find_all('h3',class_='lister-item-header')
```

```python
Name[0:4]
```

```python
for i in Name:
    Movies_Name.append(i.get_text().replace("\n",""))
Movies_Name
```

```python
Rating=soup.find_all('div',class_='inline-block ratings-imdb-rating')
Rating[0:4]
```

```python
for i in Rating:
    IMDB_Rating.append(i.get_text().replace("\n",""))
IMDB_Rating
```

```python
Release=soup.find_all('span',class_='lister-item-year text-muted unbold')
Release[0:4]
```

```python
for i in Release:
    Year_Of_Release.append(i.get_text().replace("\n",""))
Year_Of_Release
```

```python
import pandas as pd
```

```python
IMDB=pd.DataFrame({})
```

```python
IMDB['NAME']=Movies_Name
IMDB['RATING']=IMDB_Rating
IMDB['YEAR']=Year_Of_Release
```
